```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(context="paper", font="monospace")
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Distplot

Used to check the distribution of one column

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

### Jointplot

This is a bivariate distribution plot. Used to check the distribution between two columns

The most important parameters are x, y and data

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='scatter')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde')
```

### Pairplot

Plots pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

### CATEGORICAL PLOTS

Categorical plots are used for visualizing distribution of categorical data.

Examples are: factorplot/catplot, boxplot, violinplot, stripplot, swarmplot, barplot, countplot, etc

### barplot

barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean

```python
sns.barplot(x='sex',y='total_bill',data=tips)
# Note that by default, the estimator is the mean
```

```python
# You can set the estimator to the standard deviation or sum
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.sum)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

### countplot

Checks for number of occurences, just like pandas count

```python
sns.countplot(x='sex', data=tips)
```
